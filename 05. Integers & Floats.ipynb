## Understanding Numbers (Integers and Floats) in Python

Python has built-in support to store and process numeric data (Python Numbers). 

Most of the times you work with numbers in almost every Python application. Obviously, any computer application deals with numbers.

Numbers are a fundamental part of programming, and Python provides powerful tools to work with various types of numbers. Whether you need to count, calculate, or perform complex mathematical operations, Python has you covered.

## <span style = "color:green"> 1] INTEGERS </span>


Integers are whole numbers that can be positive, negative, or zero. 

They do not have a fractional or decimal component.

***Examples***: -5, 0, 49

## <span style = "color:green"> 2] FLOATS </span>


Floats (floating-point numbers) are numbers that have a decimal point.

They can represent fractions and provide more precision than integers.

***Example***: 3.14, -0.001, 2.718

In Python, integers and floats can be mixed in arithmetic operations, and the result will be a float if at least one operand is a float.


Floats are particularly useful for calculations requiring decimal precision, such as scientific computations and measurements.

#### <span style = "color:red"> Additional arithmetic tricks and operations </span>

```python
# Floor Division (//)
# Operation: Perform division and does not round the result; instead, it simply returns the integer part of the quotient by discarding the fractional part.

102//4
```

```python
102/4
```

```python
15/4
```

```python
15//4
```

```python
# Modulus (%)
# Operation: Find the remainder after division

5%2
```

```python
# Exponentiation (*)
# Operation: Raise a number to the power of another number. (Powers)

2**6
```

```python
2**3
```

```python
# Can also do roots this way (number ** 0.5)
4**0.5
```
